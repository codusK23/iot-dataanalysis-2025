### K-최근접 이웃 알고리즘
- K-Nearest Neighbors algorithm

- 도미/빙어 데이터 합치기


```python
length = bream_length + smelt_length
weight = bream_weight + smelt_length
```

```python
len(length) # 35 + 14
```

```python
len(weight)
```

```python
for l, w in zip(length, weight):
    print(l, w)
```

```python
# zip
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
print(fish_data)
```

각 데이터에 대한 분류데이터 생성

```python
# 타켓 데이터
fish_target = [1] * 35 + [0] * 14
print(fish_target)
print(len(fish_target))
```

- 도미 1, 빙어 0 으로 분류

#### K-최근접 이웃 알고리즘 사용
- 사이킷런 모듈

```python
from sklearn.neighbors import KNeighborsClassifier
```

- kn은 모델 객체

```python
kn = KNeighborsClassifier() # 모델 객체 생성
```

- fit() 함수로 `훈련(training)` 시작

```python
# X : 특성데이터, y: 결과데이터
kn.fit(fish_data, fish_target)
```

- 훈련 점수 도출

```python
# 훈련시킨 데이터를 그대로 집어넣어서 결과가 제대로 나오는지 확인
kn.score(fish_data, fish_target)
```

#### K-최근접 이웃 알고리즘
- 주어진 데이터 내에서 새로운 데이터의 근처에 있는 데이터가 무엇인지 확인해서 값을 도출하는 방법

- predict()로 예측으로 수행

```python
# 길이 30cm, 무게 600g 물고기가 어떤 종류인지 분류 예측
kn.predict([[30, 600]])
```

- 위의 데이터는 도미라고 예측

```python
# 도미/빙어 산점도
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter([30], [600], marker='^') # 예측하고자 하는 데이터1
plt.scatter([13.0], [14.2], marker='*') # 예측하고자 하는 데이터2
plt.xlabel('도미/빙어 길이')
plt.ylabel('도미/빙어 무게')
plt.title('도미/빙어 산점도')
plt.show()
```

#### 분류시 참조하는 이웃 데이터 갯수

```python
kn.n_neighbors
```

- 이웃으로 참조할 데이터 갯수를 49개로 변경

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
```
